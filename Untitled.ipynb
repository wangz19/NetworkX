```python
import networkx as nx
```

```python
import glob
glob.glob('*.txt')
```

We can read the adjacent list from txt file


```python
file_name = 'BAedges.txt'
```

```python
# read in the file
g = nx.read_edgelist(file_name, create_using = nx.Graph(),nodetype = int)
```

```python
nx.info(g)
```

```python
nx.average_shortest_path_length(g)
```

```python
nx.shortest_path_length(g,source = 0)
```

```python
a = []
for j in [i[1] for i in nx.shortest_path_length(g)]:
    a.extend(list(j.values())) 
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(a,bins = 'auto')
plt.show()
```

```python
sns.distplot(a, bins = len(set(a)), kde = False)
```
